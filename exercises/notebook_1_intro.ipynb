```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values from 0 to 2 Pi
x = np.linspace(0, 2 * np.pi, 1000)

# Compute y values as sin(x)
y = np.sin(x)

# Create the plot
plt.plot(x, y)

# Add title and labels
plt.title('Sine Function')
plt.xlabel('x')
plt.ylabel('sin(x)')

# Show the plot
plt.show()
```
